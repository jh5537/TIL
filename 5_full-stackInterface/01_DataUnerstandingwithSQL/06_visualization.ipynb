```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.

ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend() 범례 추가. (loc=)값에 따라 위치를 지정할 수 있음.
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

tips 파일 분석 종합

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
tips = pd.read_csv('tips.csv')
tips.head(5)
```

```python
tips.info()
```

```python
t_bill = tips.total_bill
t_bill.describe()
```

```python
plt.boxplot(t_bill, labels=['Total Bill'])
plt.show()
```

```python
t_bill_sns = sns.boxplot(x='total_bill', data=tips)
```

```python
t_tips = tips.tip
```

```python
plt.boxplot(t_tips, labels=['Tip'])
plt.show()
```

```python
t_tips_sns = sns.boxplot(x='tip', data=tips)
```

```python
sns.regplot(x='total_bill', y='size', data=tips)
```

```python
sns.regplot(x='size', y='tip', data=tips)
```

```python
sns.histplot(t_bill, bins=50)
```

```python
sns.distplot(t_bill, bins=50) # 해당 기능 지원 중단하여 경고 표시.
```

```python
sns.distplot(t_tips, bins=50)
```

```python
countsize = pd.crosstab(index=tips['size'], columns='count')
countsize
```

```python
countsize.plot.bar()
```

```python
size_sex = pd.crosstab(index=tips['size'], columns=tips['sex'])
size_sex
```

```python
size_sex.plot.bar()
```

```python
tip_rate = tips.tip/tips.total_bill
tips['tip_rate'] = tip_rate
tips.describe()
```

```python
tips[tips['tip_rate']>0.25]
```

```python
tips.groupby([tips['day'],tips['time']]).count()
```

```python
tips = tips.assign(BPP=tips["total_bill"] / tips["size"])
tips
```

```python
tips['sex']=tips['sex'].replace({'Female': 0, 'Male':1})
tips['smoker']=tips['smoker'].replace({'No': 0, 'Yes':1})
tips['day']=tips['day'].replace({'Thur': 0, 'Fri':1, 'Sat': 2, 'Sun': 3})
tips['time']=tips['time'].replace({'Lunch': 0, 'Dinner':1})

tips.head(3)
```

```python
import pandas_profiling
tips.profile_report()
```

```python
tips.corr()
```
